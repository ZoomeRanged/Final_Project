```python
import numpy as np
import pandas as pd 
from sqlalchemy import create_engine
```

```python
url = 'postgresql://postgres:admin@host.docker.internal:5434/postgres'
engine = create_engine(url)
```

```python
df_train = pd.read_sql_query('SELECT * FROM home_credit_default_risk_application_train_ok', con=engine)
df_train.head()
```

```python
df_test = pd.read_sql_query('SELECT * FROM home_credit_default_risk_application_test_ok', con=engine)
df_test.head()
```

```python
drop_columns = []
for col in df_train.columns:
    null_pct = df_train[col].isna().sum() / len(df_train)
    if null_pct > 0.06:
        drop_columns.append(col)

df_train_ok = df_train.drop(drop_columns, axis=1)
df_train_ok.head()
```

```python
df_test_ok = df_test.drop(drop_columns, axis=1)
df_test_ok.head()
```

```python
X_train = df_train_ok.drop(columns=['TARGET'],axis=1)
y_train = df_train_ok['TARGET']
X_test = df_test_ok
```

```python
# numerical
num = X_train.select_dtypes(exclude='object').columns

# Categorical
cat = X_train.select_dtypes(include='object').columns
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

from sklearn.linear_model import LogisticRegression
from imblearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split

from sklearn.metrics import precision_recall_fscore_support
```

```python
cat_transformer = Pipeline([
    ("c_i", SimpleImputer(strategy="most_frequent")),
    ("e", OneHotEncoder(handle_unknown = "ignore"))
])

num_transformer = Pipeline([
    ("n_i", SimpleImputer(strategy="median"))
])

transformers = [
    ("n_t", num_transformer, num),
    ("c_t", cat_transformer, cat)
]

model = Pipeline([
    ("pre", ColumnTransformer(transformers=transformers)),
    ("model", LogisticRegression(max_iter=1000, random_state=1))
])
```

```python
model.fit(X_train, y_train)
```

```python
y_test = model.predict(X_test)
```

```python
y_prob = model.predict_proba(X_test)
```

```python
ml_result = X_test[['SK_ID_CURR']]

# add ml result
ml_result['prediction_target'] = y_test.tolist()
ml_result['probability'] = y_prob.tolist()

ml_result
```

```python
ml_result.to_sql('home_credit_default_risk_application_ml_result', index=False, con=engine, if_exists='replace')
```
